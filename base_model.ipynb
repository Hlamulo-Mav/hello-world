## Import the libraries

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#Accuracy packages
from sklearn.metrics import mean_squared_error
```

## Import the dataset

## Test dataset

```python
test_set= pd.read_csv('https://raw.githubusercontent.com/thembeks/Regression-Sendy-Logistics-Challenge-Team-14/Predict/Test.csv')
test_set.select_dtypes(include=['float', 'int'])
test_set.columns= [col.replace(' ', '_') for col in test_set.columns]
test_set.head()
```

```python
#dropping outlier columns and that cannot be encoded 
test_set.drop(['Order_No', 'User_Id', 'Rider_Id', 'Vehicle_Type'], inplace= True, axis=1)
test_set.head()
```

## Train dataset

```python
#formating train dataframe 
train_set= pd.read_csv('https://raw.githubusercontent.com/thembeks/Regression-Sendy-Logistics-Challenge-Team-14/Predict/Train.csv')
train_set.columns= [col.replace(' ', '_') for col in train_set.columns]
train_set.head()
```

## Rider dataset

```python
rider= pd.read_csv('https://raw.githubusercontent.com/thembeks/Regression-Sendy-Logistics-Challenge-Team-14/Predict/Riders.csv')
rider.columns= [col.replace(' ', '_') for col in rider.columns]
rider.head()
```

## Merge Rider and Train dataset into a new dataset Train

```python
Train = pd.merge(train_set, rider, how ='left', left_on=['Rider_Id'], right_on =['Rider_Id'])
Train.head()
```

## Data cleanup

```python
#Converting time strings to seconds for the Train dataframe

Train['Placement_-_Time']= pd.to_timedelta(Train['Placement_-_Time']).dt.total_seconds()
Train['Confirmation_-_Time']= pd.to_timedelta(Train['Confirmation_-_Time']).dt.total_seconds()
Train['Arrival_at_Pickup_-_Time']= pd.to_timedelta(Train['Arrival_at_Pickup_-_Time']).dt.total_seconds()
Train['Pickup_-_Time']= pd.to_timedelta(Train['Pickup_-_Time']).dt.total_seconds()
Train.head()
```

```python
#Converting time strings to seconds for the test_set dataframe

test_set['Placement_-_Time']= pd.to_timedelta(test_set['Placement_-_Time']).dt.total_seconds()
test_set['Confirmation_-_Time']= pd.to_timedelta(test_set['Confirmation_-_Time']).dt.total_seconds()
test_set['Arrival_at_Pickup_-_Time']= pd.to_timedelta(test_set['Arrival_at_Pickup_-_Time']).dt.total_seconds()
test_set['Pickup_-_Time']= pd.to_timedelta(test_set['Pickup_-_Time']).dt.total_seconds()
test_set.head()
```

```python
#dropping outlier columns and that cannot be encoded 
Train.drop(['Order_No', 'User_Id', 'Rider_Id', 'Vehicle_Type'], inplace= True, axis=1)
Train.head()
```

## Missing data

```python
#replacing missing data with average of each column for Train dataframe
Train.fillna(Train.mean(), inplace=True)
Train.head()
```

```python
#replacing missing data with average of each column for test_set dataset
test_set.fillna(test_set.mean(), inplace=True)
test_set.head()
```

## Encoding categorical data

```python
new_df =  pd.get_dummies(Train, columns=['Platform_Type', 'Personal_or_Business'],
                   drop_first= True)
new_df.head()
```

```python
##encoding test dataframe
new_test =  pd.get_dummies(test_set, columns=['Platform_Type', 'Personal_or_Business'],
                   drop_first= True)
new_test.head()
```

## Splitting the dataset

```python
 # Making sure that the targer variable is the last column
df = new_df[['Placement_-_Day_of_Month', 'Placement_-_Weekday_(Mo_=_1)', 'Confirmation_-_Day_of_Month', 'Confirmation_-_Weekday_(Mo_=_1)', 'Arrival_at_Pickup_-_Day_of_Month', 'Arrival_at_Pickup_-_Weekday_(Mo_=_1)', 'Pickup_-_Day_of_Month',
'Pickup_-_Weekday_(Mo_=_1)', 'Arrival_at_Destination_-_Day_of_Month', 'Arrival_at_Destination_-_Weekday_(Mo_=_1)','Distance_(KM)', 'Temperature','Precipitation_in_millimeters', 'Pickup_Lat', 'Pickup_Long', 'Destination_Lat', 'Destination_Long','Platform_Type_2', 'Platform_Type_3','Platform_Type_4', 'Personal_or_Business_Personal', 'Time_from_Pickup_to_Arrival']]
```

```python
df
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
#splitting the data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```
